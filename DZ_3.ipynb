### Описание:
Использована модель - [seara/rubert-tiny2-russian-sentiment](https://huggingface.co/seara/rubert-tiny2-russian-sentiment)  
Модель RuBERT-tiny2 , настроенная для классификации по тональности коротких русских текстов. Задача представляет собой классификацию нескольких классов со следующими метками:
```
0: neutral
1: positive
2: negative
```
#### Категории для русского языка:
```
neutral: нейтральный
positive: позитивный
negative: негативный
```
#### Usage
```
from transformers import pipeline
model = pipeline(model="seara/rubert-tiny2-russian-sentiment")
model("Привет, ты мне нравишься!")
# [{'label': 'positive', 'score': 0.9398769736289978}]
```

````python
from transformers import pipeline

classifier = pipeline(model="seara/rubert-tiny2-russian-sentiment")
````

````python

emotion_labels_ru = {
    "neutral": "нейтральный",
    "positive": "позитивный",
    "negative": "негативный",
}
````

````python
def classify_texts_with_sentiments(text, classifier, dict):
    output = ""
    for i, text in enumerate(text, start=1):
        result = classifier([text])[0]
        label = result["label"]
        score = result["score"]
        output += f"{i}. Текст: {text}\n   Настроение: {dict[label]}\n   Оценка: {score}\n"
    return output
````

````python
texts = ["Отличный товар", "Мне не понравилось", "Прекрасно выглядит, ничего не меняйте!", "Ничего не понял", "У меня не работала розетка, в остальном все отлично"]
````

````python
print(classify_texts_with_sentiments(texts, classifier, emotion_labels_ru))
````
